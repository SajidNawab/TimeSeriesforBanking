## Simple Row Slicing

The simplest way to slice a `DataFrame` is to use square brackets: `[]`.  The syntax `df[i:j]` will generate a `DataFrame` who's first row is the `i`th row of `df` and who's last row is the `(j-1)`th row of `df`.

Let's demonstrate this with a couple of examples:

```python
df_hbl[10:20]
```

Starting with the 3rd row, and ending with the 6th row:

```python
df_hbl[3:7]
```

The full `DataFrame`:

```python
df_hbl # full dataframe
```

## Indexing with `DataFrame.iloc`

The indexer attribute `DataFrame.iloc` can be used to access rows and columns using their implicit row and column numbers.

Here is an example of `iloc` that retrieves the first two rows of `df_hbl`:

```python
df_hbl.iloc[1:4][['Open','Close']]
```

Notice, that because we didn't specify any particular column numbers, the code above retrieves all columns.

The following code grabs just the first three rows and the first three columns of the first two rows of `df_hbl`:

```python
r=[15]
c=[1]
df_hbl.iloc[r,c]
```

We can also supply `.iloc` with `lists` rather than ranges to specify custom sets of columns and rows:

```python
lst_row = [0, 2] # first and third rows
lst_col = [0, 6] # date and adjusted columns
df_hbl.iloc[0:10, [0,6]]
```

Using `lists` as a means of indexing is sometimes referred to as *fancy indexing*.

## Indexing with `DataFrame.loc`

Rather than using the implicit row or column numbers, it is often more useful to access data by using the explicit row or column indices.

As a preliminary step, let's first recast the `date` column as a `datetime`.  This can be done easily by using `pandas` built-in `to_datetime()` method:

```python
df_hbl['Date'] = pd.to_datetime(df_hbl['Date']) 
df_hbl.dtypes
```

```python
df_hbl['Year']=df_hbl['Date'].dt.year
df_hbl['Month']=df_hbl['Date'].dt.month
df_hbl
```

Next we'll use the `DataFrame.set_index()` method to set the `date` column as our new index.  We are doing this because we ultimately want to use our dates as an index to access the data.

```python
df_hbl.set_index('Date',inplace=True)
df_hbl.head()
```

To see the effect of the above code, we can have a look at the `index` of `df_hbl`.

```python
df_hbl.loc['2011-01-10']
# df_hbl.index
```

Notice that `date` is no longer column of `df_hbl`:

```python
df_hbl.columns
```

Now that we have successful set the row `index` of `df_hbl` to be the `date`, let's see how we can use this `index` to access the data via `.loc`.
        
Here is an example of how we can grab a slice of rows, associated with a date-range:

```python
df_hbl.loc['2018-12-21':'2018-12-28']
```

If we want to select only the `Turnover` and `Close` columns for these dates, we would type the following: 

```python
df_hbl.loc['2018-12-21':'2018-12-28', ['Turnover', 'Close']]
```

```python
# # create a line plot for each company with their Yearly turnover (summed)
import matplotlib.pyplot as plt 
plt.figure(figsize=(16,9))
df_volume = df.groupby(['company', 'year'])['Turnover'].sum().reset_index()
df_volume[df_volume['company']=='ABL']['Turnover']
years=list(df_volume['year'].unique())
abl=list(df_volume[df_volume['company']=='ABL']['Turnover'])
ubl=list(df_volume[df_volume['company']=='UBL']['Turnover'])
hbl=list(df_volume[df_volume['company']=='HBL']['Turnover'])
bop=list(df_volume[df_volume['company']=='BOP']['Turnover'])

plt.plot(years, abl,marker='o',color='g',markerfacecolor='r',linestyle='--',label='ABL')
plt.plot(years, ubl,marker='v',color='r',markerfacecolor='b',linestyle='-',label='UBL')
plt.plot(years, bop,marker='*',color='m',markerfacecolor='y',linestyle=':',label='BOP')
plt.plot(years, hbl,marker='.',color='b',markerfacecolor='m',linestyle='--',label='HBL')


plt.title("Turnover Graph")
plt.xlabel("Turnover (in Millions)")
plt.ylabel("Years")
plt.legend()
plt.show()
```
